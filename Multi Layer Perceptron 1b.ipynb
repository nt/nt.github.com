## Loading the data

````python
import numpy as np
import os

work_dir = os.environ['WORK_DIR']
x = np.load(work_dir + "/x.npy").astype('f')
y = np.load(work_dir + "/y.npy").astype('f')

validation_size = len(x)/10

# permute
c = np.concatenate((x,y), axis=1)
c = np.random.permutation(c)

xx = c[:,0:x.shape[1]]
yy = c[:,x.shape[1]:][:,0]

x_valid = xx[-validation_size:,:]
x_train = xx[:-validation_size,:][0:3000]

#yy = np.array([1-yy[:,0], yy[:,0]]).T

y_valid = yy[-validation_size:]
y_train = yy[:-validation_size][0:3000]
````

## Symbolic declaration

![Formula](http://deeplearning.net/tutorial/_images/math/f3ac04d8d5fb983a75346ecba39fc309b634f968.png)

This is a two layer network. The final layer does logistic regression so ```G``` is ```argmax```. For the inner non linear function ```s``` we use ```tanh```.

````python
import theano
import theano.tensor as T

rng = np.random.RandomState(1234)

n_in = x_train.shape[1]
hidden_size = 1500
learning_rate = 1e-2

t_x = T.fmatrix('x')
t_y = T.lvector('y')

t_b1 = theano.shared(value=np.zeros((hidden_size,), dtype=theano.config.floatX), name='b1')
t_W1 = theano.shared(value=np.asarray(rng.uniform(
        low=-np.sqrt(6. / (n_in + hidden_size)),
        high=np.sqrt(6. / (n_in + hidden_size)),
        size=(n_in, hidden_size)), dtype=theano.config.floatX), name='W1')

t_activation = T.tanh(T.dot(t_x, t_W1) + t_b1)

t_b2 = theano.shared(np.zeros((2,)), name='b2')
t_W2 = theano.shared(rng.randn(hidden_size, 2), name='W2')

t_p_y_given_x = T.nnet.softmax(T.dot(t_activation, t_W2) + t_b2)
t_y_pred = T.argmax(t_p_y_given_x, axis=1)

t_errors = T.mean(T.neq(t_y_pred, t_y))

t_L2_sqr = (t_W1 ** 2).sum() + (t_W2 ** 2).sum()

t_l2_cost = 1e-4*t_L2_sqr
t_logreg_cost = -T.mean(T.log(t_p_y_given_x)[T.arange(t_y.shape[0]), t_y])
t_cost = t_logreg_cost + t_l2_cost

t_dW1, t_db1, t_dW2, t_db2 = T.grad(t_cost, [t_W1, t_b1, t_W2, t_b2])

t_f = theano.function([t_x, t_y], [t_cost, t_logreg_cost, t_l2_cost, t_errors], updates=[
(t_b1, t_b1 - learning_rate*t_db1),
(t_W1, t_W1 - learning_rate*t_dW1),
(t_b2, t_b2 - learning_rate*t_db2),
(t_W2, t_W2 - learning_rate*t_dW2)
], allow_input_downcast=True)
````

## Training

````python
import time
def step(prin = True):
    start = time.clock()
    cost, logreg_cost, l2_cost, errors = t_f(x_train, y_train)
    if prin:
        print "Cost is ", [cost, logreg_cost, l2_cost], "errors are", errors
    elapsed = time.clock() - start
    return [cost, logreg_cost, l2_cost, errors, elapsed]
````

````python
training = []
````

````python
training += [step(i % 3 == 0) for i in xrange(35)]
````

````python
%matplotlib inline
import matplotlib.pyplot as plt

cutoff = 50
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 16))
axes[0][0].plot(np.array(training)[:,0][:cutoff])
axes[0][0].set_title("cost, begining")
axes[0][1].plot(np.array(training)[:,0][cutoff:])
axes[0][1].set_title("cost, end")
axes[1][0].plot(np.array(training)[:,2][:cutoff])
axes[1][0].set_title("l2 cost, begining")
axes[1][1].plot(np.array(training)[:,2][cutoff:])
axes[1][1].set_title("l2 cost, end")
axes[2][0].plot(np.array(training)[:,1][:cutoff])
axes[2][0].set_title("reg cost, begining")
axes[2][1].plot(np.array(training)[:,1][cutoff:])
axes[2][1].set_title("reg cost, end")
axes[3][0].plot(np.array(training)[:,3][:cutoff])
axes[3][0].set_title("error, begining")
axes[3][1].plot(np.array(training)[:,3][cutoff:])
axes[3][1].set_title("error, end")
axes[4][0].plot(np.array(training)[:,4][:cutoff])
axes[4][0].set_title("time for step, begining")
axes[4][1].plot(np.array(training)[:,4][cutoff:])
axes[4][1].set_title("time for step, end")
````
